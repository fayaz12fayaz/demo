```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("bmw_global_sales_2018_2025.csv")

df.sample(5)
df.shape
df.describe()
df.info()
df.isnull().sum()
df.columns
df
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(df.iloc[:,:4],df.iloc[:,4],test_size=0.2)


x_train["Model"].value_counts()
x_train["Region"].value_counts()

# x_train["Month"].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

transformer = ColumnTransformer(transformers=[
    ("tns1",OneHotEncoder(sparse_output=False,drop="first"),["Region","Model"])
],remainder="passthrough")

x_train_transformed = transformer.fit_transform(x_train)
x_test_transformed = transformer.transform(x_test)

x_test_transformed
```
